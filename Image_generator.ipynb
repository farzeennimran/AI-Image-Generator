```python
from huggingface_hub import notebook_login
from diffusers import StableDiffusionPipeline

# Login to Hugging Face
notebook_login()  # This will prompt you to enter your API key
```

```python
import os

# Set your Hugging Face API token
os.environ['HUGGINGFACE_TOKEN'] = 'YOUR API KEY'
```

```python
pipe = StableDiffusionPipeline.from_pretrained(
    "CompVis/stable-diffusion-v1-4",
    use_auth_token=True
)
pipe = pipe.to("cuda")  # If using GPU
```

```python
prompt = "red full moon shown from earth"
image = pipe(prompt).images[0]

# Display the image
image.show()

# Save the image
image.save("moon_from_earth.png")
```

```python
pipe = StableDiffusionPipeline.from_pretrained(
    "CompVis/stable-diffusion-v1-4",
    use_auth_token=True,
    revision="fp16",
    torch_dtype=torch.float16
)
pipe = pipe.to("cuda")

image = pipe(prompt, num_inference_steps=50, guidance_scale=7.5).images[0]
```

```python
import gradio as gr

def generate_image(prompt):
    image = pipe(prompt).images[0]
    return image

gr.Interface(fn=generate_image, inputs="text", outputs="image").launch()
```
